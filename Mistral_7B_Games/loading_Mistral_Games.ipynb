```python
# import libraries

import os
import numpy as np
import pandas as pd
import urllib.parse
from urllib.parse import unquote
```

```python
import pickle
with open('Mistral_Games.pickle', 'rb') as file:
    Mistral_Games = pickle.load(file)
```

```python
df = pd.DataFrame.from_dict(Mistral_Games, orient='index')
df.head()
```

```python
list_of_lists = []
for key in Mistral_Games:
    for i in range(len(Mistral_Games[key])):
        list_of_lists.append([key, Mistral_Games[key][f'rep{i}'], i])

df = pd.DataFrame(list_of_lists, columns=['game', 'results', 'rep'])
df
```
